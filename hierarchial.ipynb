```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, -2:].values
print(X)
```

```python
from scipy.cluster import hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
y_agglomerative = cluster.fit_predict(X)
print(y_agglomerative)
```

```python
plt.scatter(X[y_agglomerative==0, 0],  X[y_agglomerative==0, 1], s = 100, label='Cluster 1')
plt.scatter(X[y_agglomerative==1, 0],  X[y_agglomerative==1, 1], s = 100, label='Cluster 2')
plt.scatter(X[y_agglomerative==2, 0],  X[y_agglomerative==2, 1], s = 100, label='Cluster 3')
plt.scatter(X[y_agglomerative==3, 0],  X[y_agglomerative==3, 1], s = 100, label='Cluster 4')
plt.scatter(X[y_agglomerative==4, 0],  X[y_agglomerative==4, 1], s = 100, label='Cluster 5')
plt.legend()
plt.title('Hierarchial Clusters of Customers')
plt.ylabel('Spending Score (1-100)')
plt.xlabel('Annual Income (thousands $)')
plt.show()
```
